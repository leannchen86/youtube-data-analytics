# Filter comments containing specific words

```python
all_chnl_comments_df['date']=pd.to_datetime(all_chnl_comments_df['date'], format="%Y/%m/%d").dt.date
```

```python
def filtered(cmt):
    keyword_lst = tuple(['great', 'excellent', 'outstanding', 'brilliant', 'knowledge', 'respect', 'subscribed', 'masterpiece',\
               'impressed', 'story', 'storytelling', 'informational', 'informative', 'entertaining', 'appreciated',\
               'amazing', 'awesome', 'gifted', 'storyteller', 'beautiful', 'wonderful', 'perfect', 'insight', 'best',\
               'fan', 'epic', 'love', 'genius', 'wow', 'thank', 'treasure', 'passion', 'pleasure', 'inform', 'educat',\
               'unreal', 'incredible', 'quality', 'well done', 'magnificent', 'master', 'appreciate', 'breakdown', 'kenshin',\
                'lawrence', 'expert', 'matrix']) 
    for w in keyword_lst:
        if w in cmt:
            return True
        else:
            if w != keyword_lst[-1]:
                continue
            else:
                return False
```

```python
all_chnl_comments_df['comment'] = all_chnl_comments_df['comment'].map(lambda x: x.lower())
all_chnl_comments_df['value'] = all_chnl_comments_df['comment'].apply(lambda x: filtered(x))
```

```python
positive_comments_df = all_chnl_comments_df[all_chnl_comments_df['value']==True]
```

```python
# Do not include channel owner's comments
positive_comments_df = positive_comments_df[positive_comments_df['author_id']!='UC9tDOUA6AnQ7LNe3b70QwKA']
```

```python
# data-cleaning: drop redundant columns
positive_comments_df.drop(['comment_id', 'video_id', 'value'], axis=1, inplace=True)
```

```python
# sort by likes
positive_cmt_likes = positive_comments_df.sort_values(by='likes', ascending=False)
# sort by replies
positive_cmt_replies = positive_comments_df.sort_values(by='replies', ascending=False)
```
